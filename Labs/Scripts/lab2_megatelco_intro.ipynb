**I will use the dropna function to clean the data to remove rows with missing values**

```python
print("Before dropping rows with missing values:")
print(df_clean.shape)
```

```python
df_clean = df_clean.dropna()
print("\nAfter dropping rows with missing values:")
print(df_clean.shape)
```



---



** I want to further invisitgate and filter the data to include only rows where the reported satisfaction is either 'low' or 'high' that excludes rows with other values.**


```python
# Before filtering based on reported satisfaction
print("Before filtering based on reported satisfaction:")
print(df_clean.shape)

```

```python
allowed_satisfaction_values = ['low', 'high']
df_clean = df_clean[df_clean['reported_satisfaction'].isin(allowed_satisfaction_values)]

print("\nAfter filtering based on reported satisfaction:")
print(df_clean.shape)
```



---



**I further wanted to filter the data to include only rows where customers are considering a change of plan.**

```python
print("Before filtering based on considering change of plan:")
print(df_clean.shape)

```

```python
df_clean = df_clean[df_clean['considering_change_of_plan'] == 'yes']

print("\nAfter filtering based on considering change of plan:")
print(df_clean.shape)
```
